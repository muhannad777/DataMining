```python
df_with_delimiter=pd.read_csv(r'D:\\python\lib\site-packages\sklearn\datasets\data\iris.csv', delimiter=',', header= None,
                             names=['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)','Target Class'])

df_with_delimiter.tail()
```

```python
for ojha, feature in enumerate(list(df_with_delimiter.columns)[:-1]):
    
    fg = sns.FacetGrid(df_with_delimiter, hue='Target Class', height=5)
    
    fg.map(sns.distplot, feature).add_legend()
    
    plt.show()
```

```python
sns.boxplot(x='Target Class',y='petal length (cm)', data=df_with_delimiter)
plt.show()
```

```python
sns.violinplot(x='Target Class',y='petal length (cm)', data=df_with_delimiter)
plt.show()
```

```python
df_with_delimiter.plot(kind='scatter',x='sepal length (cm)',y='sepal width (cm)', c= 'g')

plt.show()
```

```python
sns.set_style("whitegrid");

sns.pairplot(df_with_delimiter,vars = ['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)'], hue = 'Target Class', diag_kind = 'kde', 

       plot_kws = {'alpha': 0.6, 's': 80, 'edgecolor': 'k'}, height = 2) 

plt.show()
```
